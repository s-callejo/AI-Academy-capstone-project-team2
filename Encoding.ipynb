```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#read
df = pd.read_csv("E-commerce Website Logs new.csv")
df.head()
```

```python
#start prep
df.drop(['network_protocol', 'ip', 'bytes'], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df = df[df['age']!="--"] #remove non-integer values
```

```python
#turn age into integer
df['age'] = df['age'].astype('int64')
```

```python
df = df[df['age']!=0]
```

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df['returned'] = le.fit_transform(df['returned'])
```

```python
df['membership'].value_counts()
```

```python
encoded_columns = ['accessed_Ffom', 'country', 'membership', 'language', 'pay_method']
ohe = pd.get_dummies(df[encoded_columns],prefix=encoded_columns, drop_first=True)
df_encoded = pd.concat([df, ohe], axis=1)
df_encoded = df_encoded.drop(encoded_columns,axis=1)
#df['accessed_Ffom'] = le.fit_transform(df['accessed_Ffom'])
#df['country'] = le.fit_transform(df['country'])
#df['membership'] = le.fit_transform(df['membership'])
#df['language'] = le.fit_transform(df['language'])
#df['pay_method'] = le.fit_transform(df['pay_method'])
```
